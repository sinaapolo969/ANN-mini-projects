```python
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def plot_distribution(function, name):

    fig, ax = plt.subplots(1, 3)
    fig.set_size_inches(w = 12, h = 4)
    
    #plot function
    x_values = np.linspace(-10, 10, 100)
    y_values = function(x_values)
    ax[0].plot(x_values, y_values, label = name)
    ax[0].legend()
    input = np.random.randn(10000)
    #plot input
    ax[1].hist(input, label = 'Input distribution')
    ax[1].legend()

    #plot output
    out = function(input)
    ax[2].hist(out, label = f'distribution of {name}')
    ax[2].legend()
```

# 1- Sigmoid

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
def d_sigmoid(x):
    return sigmoid(x) * (1 - sigmoid(x))
```

```python
plot_distribution(sigmoid, 'sigmoid')
```

```python
# derivative of sigmoid
plot_distribution(d_sigmoid, 'derivative of sigmoid')
```

# identity function 

```python
def identity(x):
    return x
def d_identity(x):
    return np.ones(len(x))
```

```python
plot_distribution(identity, 'identity')
```

```python
# derivative of identity 
plot_distribution(d_identity, 'derivative of identity')
```

# 3- Relu

```python
def relu(x):
    return np.where(x > 0, x, 0)
def d_relu(x):
    return np.where(x > 0, 1, 0)
```

```python
plot_distribution(relu, 'ReLU')
```

```python
# d_relu
plot_distribution(d_relu, 'derivativi of ReLU')
```

# 4- Leaky ReLU 

```python
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)
def d_leaky_relu(x, alpha = 0.01):
    return np.where(x > 0, 1, alpha)
```

```python
plot_distribution(leaky_relu, 'leaky_ReLU')
```

```python
plot_distribution(d_leaky_relu, 'd of leaky ReLU')
```

# 5- Gelu

```python
def gelu(x):
    return x * 0.5 * (1 + np.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * np.power(x, 3))))

def d_gelu(x):
    return 0.5 * (1 + np.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * np.power(x, 3)))) + x * (1 - np.square(np.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * np.power(x, 3))))) * np.sqrt(2 / np.pi) * (0.5 * (x + 0.044715 * np.power(x, 3)) + 0.5 * (1 + np.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * np.power(x, 3)))))
```

```python
plot_distribution(gelu, 'GeLU')
```

```python
plot_distribution(d_gelu, 'd of GeLU')
```

# 6- Softplus 

```python
def softplus(x):
    return np.log(1 + np.exp(x))

def d_softplus(x):
    return np.exp(x) / (1 + np.exp(x))
```

```python
plot_distribution(softplus, 'softplus')
```

```python
plot_distribution(d_softplus, 'd of softplus')
```

# 7- Tanh

```python
def tanh(x):
    return np.tanh(x)
def d_tanh(x):
    return 1 - np.tanh(x)**2
```

```python
plot_distribution(tanh, 'tanh')
```

```python
plot_distribution(d_tanh, 'd of tanh')
```

# 8- Swish

```python
def swish(x, beta=0.5):
    return x * (1 / (1 + np.exp(-beta * x)))

def d_swish(x, beta=0.5):
    sigmoid_beta_x = 1 / (1 + np.exp(-beta * x))
    return sigmoid_beta_x + x * beta * sigmoid_beta_x * (1 - sigmoid_beta_x)
```

```python
plot_distribution(swish, 'swish')
```

```python
plot_distribution(d_swish, 'd of swish')
```

# 9- Arctan 

```python
def arctan(x):
    return np.arctan(x)

def d_arctan(x):
    return 1 / (1 + x**2)
```

```python
plot_distribution(arctan, 'arctan')
```

```python
plot_distribution(d_arctan, 'd of arctan')
```

# 10 - Step function 

```python
def step_function(x):
    return np.where(x < 0, 0, 1)

def d_step_function(x):
    return np.where(x == 0, np.inf, 0)
```

```python
plot_distribution(step_function, 'step_function')
```

```python
plot_distribution(d_step_function, 'd of step function')
```
